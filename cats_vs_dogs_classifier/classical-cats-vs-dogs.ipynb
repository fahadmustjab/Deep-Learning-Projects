```python
train_filenames = os.listdir('./train')
def make_dataset(filenames):
    categories = []
    for f_name in filenames:
        category=f_name.split('.')[0]
        if category == "dog":
            categories.append('dog')

        else:
            categories.append('cat')
    df = pd.DataFrame({'filename':filenames,'category':categories})
    return df
```

```python
df = make_dataset(train_filenames)
```

```python
df.sample(200)
```

```python
Path.Base_Path = path
```

```python
os.listdir(path)
```

```python
path = './train'
```

```python
def get_x(r): return path + '/' + r['filename']
def get_y(r): return r['category']
```

```python
dblock = DataBlock(blocks=(ImageBlock,CategoryBlock),get_x=get_x,get_y=get_y,item_tfms=RandomResizedCrop(128,min_scale=0.03),
                      batch_tfms=aug_transforms())
```

```python
dls = dblock.dataloaders(df)
```

```python
dls.train.show_batch(nrows=1,ncols=2)
```

```python
dls.train.vocab
```

```python
from fastai.callback.all import *
```

```python
learn = cnn_learner(dls,resnet50,metrics=error_rate)
```

```python
learn.lr_find()
```

```python
lr_min=3.981071586167673e-06
```

```python
learn.fit_one_cycle(7, lr_max=slice(lr_min,0.004786))
```

```python
interip = ClassificationInterpretation.from_learner(learn)
interip.plot_confusion_matrix(figsize=(12,12),dpi=60)
```

```python
interip.most_confused(min_val=5)
```

```python
interip.plot_top_losses(30)
```
